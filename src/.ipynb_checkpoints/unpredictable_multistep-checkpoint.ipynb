```python
def get_lorenz():
    def runge_kutta(time_steps, y0, system, params):
        ys = [y0]
        for t in range(len(time_steps)-1):
            dt = time_steps[t+1]-time_steps[t]
            t0 = time_steps[t]
            t1 = time_steps[t+1]
            k1 = system(t0, y0, params)
            k2 = system(t0 + dt/2, y0 + dt / 2 * k1, params)
            k3 = system(t0 + dt/2, y0 + dt / 2 * k2, params)
            k4 = system(t1, y0 + dt * k3, params)
            y0  = y0 + dt / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
            ys.append(y0)
        return np.array(ys)

    def lorentz_ode(t, xyz, params):
        x, y, z = xyz
        σ = params['σ']
        ρ = params['ρ']
        β = params['β']

        dx = σ * (y - x)
        dy = x * (ρ - z) - y
        dz = x * y - β * z

        return np.array([dx, dy, dz])

    time_steps = np.arange(0, 20000, 0.1)
    params = {'σ' : 10., 'ρ' : 28., 'β' : 8/3}
    xyz0 = np.array([1., 1., 1.])
    lorenz_solution = runge_kutta(time_steps, xyz0, lorentz_ode, params)
    x, y, z = lorenz_solution[2000:].T
    x = (x - min(x)) / (max(x) - min(x))
    return x
```

```python
N = 20
def get_counts(values, N = N):
    if len(values) == 0:
        return [0] * N
    intervals = np.linspace(0, 1, N + 1)
    result = np.zeros(N)
    
    for i in range (N):
        result[i] = np.sum((values >= intervals[i]) & (values < intervals[i + 1])) / len(values)
        
    return list(result)

def iqr(arr) -> float:
    if (len(arr)) == 0:
        return 0
    q75, q25 = np.percentile(arr, [75 ,25])
    IQR = q75 - q25
    return IQR

def kurtosis_safe(values):
        if np.unique(values).size <= 1:
            return 0
        return kurtosis(values)
```

```python
df = []
for i in range (len(indexes) - 1):
    for j in range (indexes[i] - 5000, indexes[i + 1] - 5000):
        current = get_counts(values[j])
        current.append(abs(X[5000 + j] - means[j]) > 0.05)
        current.append(entropy(values[j]))
        current.append(np.std(values[j]))
        current.append(iqr(values[j]))
        current.append(kurtosis_safe(values[j]))
        current.append(j - (indexes[i] - 5000))
        
        df.append(current)
    
```

```python
df = pd.DataFrame(df)
df = df.rename(columns={N : 'error', N + 1 : 'entropy', N + 2 : 'std', N + 3 : 'iqr', N + 4 : 'kurtosis', N + 5 : 'step'})
df.columns = df.columns.astype(str)
```

```python
df.head()
```

```python
X = df.drop(columns='error')
y = df['error']

models = {}
fbeta_scores = {}

os.makedirs('../models', exist_ok=True)

for step in range(120):
    step_data = df[df['step'] == step]
    X_step = step_data.drop(columns=['error', 'step'])
    y_step = step_data['error']
    
    class_counts = y_step.value_counts()
    if len(class_counts) < 2:
        print(f"Step {step}: Недостаточно данных для обучения (классы: {class_counts.to_dict()}). Пропускаем...")
        continue
    
    imputer = SimpleImputer(strategy='mean')
    X_step = pd.DataFrame(imputer.fit_transform(X_step), columns=X_step.columns)

    try:
        X_train, X_test, y_train, y_test = train_test_split(
            X_step, y_step, test_size=0.2, stratify=y_step, random_state=42
        )
    except ValueError as e:
        print(f"Step {step}: Ошибка при разделении данных ({e}). Пропускаем...")
        continue
    
    if len(set(y_train)) < 2 or len(set(y_test)) < 2:
        print(f"Step {step}: В обучающей или тестовой выборке отсутствует один из классов. Пропускаем...")
        continue
    
    model = GradientBoostingClassifier()
    model.fit(X_train, y_train)
    
    preds = model.predict(X_test)
    
    fbeta = fbeta_score(y_test, preds, beta=1)
    
    models[step] = model
    fbeta_scores[step] = fbeta
    
    model_filename = f'../models/model_step_{step}.joblib'
    dump(model, model_filename)
    print(f"Модель для step {step} сохранена в файл: {model_filename}")

for step, score in fbeta_scores.items():
    print(f"Step: {step}, F1-score: {score:.4f}")
```
