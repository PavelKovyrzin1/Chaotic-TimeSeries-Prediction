```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from joblib import dump, load
from itertools import product
from scipy.stats import kurtosis, entropy
from typing import Dict, List, Tuple
```

```python
def get_lorentz():
    def runge_kutta(time_steps, y0, system, params):
        ys = [y0]
        for t in range(len(time_steps)-1):
            dt = time_steps[t+1]-time_steps[t]
            t0 = time_steps[t]
            t1 = time_steps[t+1]
            k1 = system(t0, y0, params)
            k2 = system(t0 + dt/2, y0 + dt / 2 * k1, params)
            k3 = system(t0 + dt/2, y0 + dt / 2 * k2, params)
            k4 = system(t1, y0 + dt * k3, params)
            y0  = y0 + dt / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
            ys.append(y0)
        return np.array(ys)

    def lorentz_ode(t, xyz, params):
        x, y, z = xyz
        σ = params['σ']
        ρ = params['ρ']
        β = params['β']

        dx = σ * (y - x)
        dy = x * (ρ - z) - y
        dz = x * y - β * z

        return np.array([dx, dy, dz])

    time_steps = np.arange(0, 20000, 0.1)
    params = {'σ' : 10., 'ρ' : 28., 'β' : 8/3}
    xyz0 = np.array([1., 1., 1.])
    lorenz_solution = runge_kutta(time_steps, xyz0, lorentz_ode, params)
    x, y, z = lorenz_solution[2000:].T
    x = (x - min(x)) / (max(x) - min(x))
    return x
```

```python
x = get_lorentz()[100000:]
```

```python
class Simple_model:
    def __init__(self, data: np.array):
        self.data: np.array = data
        self.patterns: np.array = None
        self.motifs: Dict[np.ndarray, np.ndarray] = {}
        self.models: Dict = {}
        for step in range(120):
            self.models[step] = load(f'../models/model_step_{step}.joblib')

        self.trained = False
        
    def generate_patterns(self, L: int, K: int) -> None:
        indices = product(range(1, K + 1), repeat=L)
        patterns = np.array(list(indices))
        patterns = np.insert(patterns, 0, 0, axis=1)

        self.patterns = np.cumsum(patterns, axis=1)
        
    def get_motifs(self, pattern) -> np.array:
        n = len(self.data)
        indices = np.arange(n)

        shifted_indices = indices[:, np.newaxis] + pattern
        condition = shifted_indices <= (n - 1)
        valid_indices = np.all(condition, axis=1)
        shifted_indices = shifted_indices[valid_indices]

        motifs = np.take(self.data, shifted_indices)
        return motifs
    
    def train(self, L: int = 4, K: int = 10) -> None:
        self.generate_patterns(L, K)
        for pattern in self.patterns:
            self.motifs[tuple(pattern)] = self.get_motifs(pattern)
        self.trained = True

    @staticmethod
    def get_counts(values, N = 20):
        intervals = np.linspace(0, 1, N + 1)
        result = np.zeros(N)
        
        for i in range (N):
            result[i] = np.sum((values >= intervals[i]) & (values < intervals[i + 1])) / len(values)

        return list(result)

    @staticmethod
    def iqr(arr) -> float:
        q75, q25 = np.percentile(arr, [75 ,25])
        IQR = q75 - q25
        return IQR

    @staticmethod
    def kurtosis_safe(values):
        if np.unique(values).size <= 1:
            return 0
        return kurtosis(values)

    def predict_error(self, values: np.array, ind: int) -> bool:
        counts = [self.get_counts(values)]
        counts[0].append(entropy(values))
        counts[0].append(np.std(values))
        counts[0].append(self.iqr(values))
        counts[0].append(self.kurtosis_safe(values))
    
        expected_columns = [str(i) for i in range(20)] + ['entropy', 'std', 'iqr', 'kurtosis']
        counts = pd.DataFrame([counts[0]], columns=expected_columns)

        prediction = self.models[ind - len(self.data)].predict(counts)
        return prediction

    def process_values(self, values: np.array, ind: int) -> float or np.nan:
        if len(values) == 0:
            return np.nan
        
        if self.predict_error(values, ind):
             return np.nan
        
        return np.mean(values)

    def single_pred(self, ind: int) -> float:
        values = []
        for pattern, motif in self.motifs.items():
            begin = ind - pattern[-1]
            z_vector = self.result[begin + pattern]
            
            if np.any(np.isnan(z_vector)):
                continue

            motif_subset = motif[:, :-1]
            z_subset = z_vector[:-1]

            distances = np.linalg.norm(motif_subset - z_subset, axis=1)
            matching_indices = np.where(distances < self.epsilon)[0]

            matching_elements = motif[matching_indices]
            values.extend(matching_elements[:, -1])
        return self.process_values(np.array(values), ind)


    def predict(self, n_preds: int = 100, epsilon: float = 0.01, multistep: bool = True, preds: np.array = None) -> np.array:
        assert self.trained
        self.epsilon = epsilon
        if multistep:
            self.result = np.append(self.data, np.zeros(n_preds))
            for i in range (len(self.data), len(self.data) + n_preds):
                self.result[i] = self.single_pred(i)
            return self.result[len(self.data): len(self.data) + n_preds]
                
        else:
            self.result = np.append(self.data, preds)
            self.answer = []
            for i in range(len(self.data), len(self.data) + n_preds):
                self.answer.append(self.single_pred(i))
            return np.array(self.answer)
```

```python
len_train = 5000
n_preds = 100
```

```python
model = Simple_model(x[:len_train])
```

```python
%%time
model.train()
```

```python
%%time
result = model.predict(n_preds = n_preds)
```

```python
valid_mask = ~np.isnan(result)
mse = np.mean((result[valid_mask] - x[len_train:len_train + n_preds][valid_mask]) ** 2)
skipped = n_preds - sum(result > -100)
```

```python
plt.figure(figsize=(15, 7))
plt.xticks(fontsize=15, fontname='Arial')
plt.yticks(fontsize=15, fontname='Arial')

plt.plot(result, label='predictions', color='red')

plt.scatter(range(len(result)), result, color='red', label='prediction points')

plt.plot(x[len_train:len_train + n_preds], label='actual', color='blue')

plt.legend(fontsize=20)
plt.xlabel('T')

plt.title(f't = {len_train}, h = {n_preds}, MSE = {mse:.4f}, skipped = {skipped}', fontsize=30)
plt.show()
```
